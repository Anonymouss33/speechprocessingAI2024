Take a portion of your recorded signal which represents a vowel sound. Perform FFT on the signal snippet and observe the amplitude spectrum. Repeat the same for a few vowel sounds.

For vowel 'A'

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
from IPython.display import Audio

# Load the MP3 audio file
audio_file = "/content/audioai.unknown"
audio_signal, sample_rate = librosa.load(audio_file)

# Choose a portion of the signal (start_time and end_time are in seconds)
start_time = 1.14
end_time = 1.77
start_index = int(start_time * sample_rate)
end_index = int(end_time * sample_rate)
signal_snippet = audio_signal[start_index:end_index]

# Perform FFT
fft_result = np.fft.fft(signal_snippet)

# Define a cutoff frequency for filtering
cutoff_freq = 1000

# Create a window for filtering
window = np.zeros_like(fft_result)
num_freq_bins = len(fft_result)
window[:cutoff_freq] = 10

# Apply the window to the FFT result
filtered_fft_result = fft_result * window

# Perform inverse FFT to obtain filtered signal
filtered_signal = np.fft.ifft(filtered_fft_result).real

# Play the filtered audio segment
Audio(data=filtered_signal, rate=sample_rate)
```

```python
# Plot the filtered audio segment
plt.figure(figsize=(10, 6))
time_axis = np.linspace(start_time, end_time, len(filtered_signal))
plt.plot(time_axis, filtered_signal)
plt.title('Filtered Audio Segment')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

Spectrogram

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
import librosa

# Load the MP3 audio file
audio_file = "/content/audioai.unknown"
audio_signal, sample_rate = librosa.load(audio_file)

# Choose a portion of the signal (start_time and end_time are in seconds)
start_time = 1
end_time = 2
start_index = int(start_time * sample_rate)
end_index = int(end_time * sample_rate)
signal_snippet = audio_signal[start_index:end_index]

# Compute the spectrogram
spectrogram = librosa.feature.melspectrogram(y=signal_snippet, sr=sample_rate)

# Convert to decibels (log scale)
spectrogram_db = librosa.power_to_db(spectrogram, ref=np.max)

# Plot the spectrogram
plt.figure(figsize=(10, 6))
librosa.display.specshow(spectrogram_db, sr=sample_rate, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram of Audio Segment')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.tight_layout()
plt.show()
```
